```python
import warnings
warnings.filterwarnings('ignore')
from IPython.display import Image
import numpy as np
import pandas as pd
```

```python
# Pew Research Center는 세곙의 여론 조사, 인구 통계 등 다양한 사회 과학 연구를 수행하는 곳이다.
# Pew Research Center에서 조사한 '미국의 소득과 종교' 데이터를 사용한다.
pew = pd.read_csv('./data/pew.csv')
pew.head()
```

melt 메소드  
데이터프레임을 깔끔한 데이터로 정리하는데 사용하는 melt() 메소드는 지정한 열의 데이터를 모두 행으로 정리해준다.  
melt(데이터프레임, id_vars[, value_vars, var_name, value_name])  
id_vars    => 위치를 그대로 유지할 열 이름을 지정한다.  
value_vars => 행으로 위치를 변경할 열 이름을 지정한다.  
var_name   => value_vars로 위치를 변경한 열 이름을 지정한다.  
value_name => var_name으로 위치를 변경한 열의 데이터를 지정한 열 이름을 지정한다.

```python
# 소득 정보가 열을 구성하고 있다. => 소득 정보 열을 행 데이터로 옮기려 한다.
# id_vars 속성으로 지정한 열(religion)을 제외한 나머지 소득 정보열(<$10k, $10-20k, ..., >150k, Don't know/refused)이
# variable 열로 소득 정보열의 데이터는 value로 정리가 된다.
# value_vars 열을 생략하면 모든 열을 대상으로 메소드가 실행된다.
# 이러한 과정을 'religion' 열을 고정해여 피벗했다. 라고 말한다.
per_long = pd.melt(pew, id_vars='religion')
per_long
```

```python
# value_vars 속성을 지정하면 지정된 열만 대상으로 피벗한다.
per_long = pd.melt(pew, id_vars='religion', value_vars='$10-20k')
per_long
```

```python
# 피벗할 열이 2개 이상이면 피벗할 열을 리스트로 묶어서 value_vars 속성으로 전달하면 된다.
per_long = pd.melt(pew, id_vars='religion', value_vars=['$10-20k', '$100-150k'])
per_long
```

```python
# var_name 속성을 이용해서 variable로 표시되던 열 이름을 '최저임금' value_name 속성을 이용해서 value로 표시되던 열 이름을
# '인원수'로 변경했다.
per_long = pd.melt(pew, id_vars='religion', value_vars='$10-20k', var_name='최저임금', value_name='인원수')
per_long
```

```python
ebola = pd.read_csv('./data/country_timeseries.csv')
ebola.iloc[:, [0, 1, 2, 3, 10, 11]]
```

```python
# Date와 Day를 고정하고 나머지를 행으로 피벗하면 각 나라별 환자 수와 사망자 수를 볼 수 있다.
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
ebola_long
```

```python
# split() 메소드에 '_'를 구분자로 전달하면 Cases_Guinea를 Cases와 Guinea로 분리할 수 있다.
print(type(ebola_long.variable))
print(type(ebola_long['variable']))
print(type(ebola_long.variable.str)) # Series는 문자열 메소드를 사용할 수 없으므로 str을 사용해 문자열로 변환한다.
variable_split = ebola_long.variable.str.split('_')
# 분리된 데이터 전체는 시리즈 타입의 데이터이다.
print(type(variable_split))
print(variable_split)
# 분리된 데이터에서 특정 위치의 데이터만 얻어오면 리스트 타입의 데이터이다.
print(type(variable_split[0]))
print(variable_split[0])
print(variable_split[0][1])
```

```python
status_value = variable_split.str[0]
print(status_value.head())
status_value = variable_split.str.get(0)
print(status_value.tail())
```

```python
country_values = variable_split.str[1]
print(country_values.head())
country_values = variable_split.str.get(1)
print(country_values[-5:])
```

```python
print(ebola_long.columns)
ebola_long['status'] = status_value
ebola_long['country'] = country_values
ebola_long
```

```python
variable_split = ebola_long.variable.str.split('_')
print(type(variable_split))
# split() 메소드에 expand=True 옵션을 지정하면 데이터프레임으로 나눈다.
variable_split = ebola_long.variable.str.split('_', expand=True)
print(type(variable_split))
variable_split
```

```python
variable_split.columns = ['status', 'country']
variable_split
```

```python
ebola_parsed = pd.concat([ebola_long, variable_split], axis=1)
ebola_parsed
```
