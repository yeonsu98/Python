전국 평균 분양 가격

```python
# pd.options.display.float_format을 사용해서 데이터프레임에 저장된 실수 데이터의 출력 서식을 지정할 수 있다.
# 출력 서식 '{:,.1f}'는 천 단위마다. ','를 출력하고 '.' 뒤의 숫자만큼 소수점 아래로 표현한다.
# 'f'를 붙이지 않으면 실수가 지수 형태로 표현된다.
pd.options.display.float_format = '{:,.1f}'.format
```

```python
# groupby 메소드와 피벗 테이블(T 옵션) 활용하기
# describe() 메소드로 데이터프레임에 저장된 숫자 데이터의 요약 통계량을 확인할 수 있다.
pre_sale.groupby(pre_sale['연도']).describe()
```

```python
# describe() 메소드 실행 결과에 'T'속성을 지정하면 요약 통계량이 수직 방향으로 출력된다.
pre_sale.groupby(pre_sale['연도']).describe().T
```

```python
# 데이터프레임의 특정 열을 지정할 때 데이터프레임['열이름'] 형태로 사용하는 방식과 데이터프레임.열이름 형태로 사용하는 방식은
# 같은 기능이 실행된다.
pre_sale.groupby(pre_sale.연도).describe().T
```

전국 규모별 평균 분양 가격

```python
# pivot_table(values, index, columns[, aggfunc, fill_value]) => aggfunc, fill_value는 생략 가능하다.
# values     => 통계 함수를 적용할 데이터프레임의 열을 지정한다.
# index      => 그룹화의 첫 번째 기준으로 사용되는 피벗 테이블로 가져올 데이터프레임의 열을 지정한다.
# columns    => 그룹화의 두 번째 기준으로 사용되는 피벗 테이블로 가져올 데이터프레임의 열을 지정한다.
# aggfunc    => index와 columns로 그룹화된 values에 적용할 함수를 지정한다. => 기본값은 평균을 계산하는 함수 mean 이다.
# fill_value => NaN을 대체할 데이터를 지정한다. => 주로 0을 사용한다.
pre_sale.pivot_table('평당분양가격', '규모구분', '연도')
```

전국 지역별 평균 분양 가격

```python
# 규모 구분이 전체로 되엉있는 금액으로만 연도별 변동 금액을 살펴보자
# 규모 구분이 전체인 데이터만 추출한다.
region_year_all = pre_sale[pre_sale['규모구분'] == '전체']
region_year_all
```

```python
# 피벗 테이블 결과는 2번째 인수(index)가 피벗 테이블의 index로 사용되므로 index를 다시 지정하려면 reset_index() 메소드를
# 사용해서 다시 index를 붙여주면 된다.
# region_year = region_year_all.pivot_table('평당분양가격', '지역명', '연도')
region_year = region_year_all.pivot_table('평당분양가격', '지역명', '연도').reset_index()
region_year
```

 전국 지역별 평균 분양 가격 변동 금액


```python
region_year['변동액'] = region_year['2018'] - region_year['2015']
region_year
```

```python
max_delta_price = np.max(region_year['변동액'])
# max_delta_price
print('2015년부터 2018년까지 분양가는 계속 상승했으며, 상승액이 가장 큰 지역은 제주로 상승액은 평당 {0:,.0f}원 이다.'.\
      format(max_delta_price))
```

```python
min_delta_price = np.min(region_year['변동액'])
min_delta_price
```

```python
mean_delta_price = np.mean(region_year['변동액'])
mean_delta_price
```
