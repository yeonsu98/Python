```python
import warnings
warnings.filterwarnings('ignore')
from IPython.display import Image
import numpy as np

def pprint(arr):
    print('type : {}'.format(type(arr)))
    print('shape : {}, dimension : {}, dtype : {}'.format(arr.shape, arr.ndim, arr.dtype))
    print('numpy 배열에 저장된 데이터\n', arr)
```

인덱싱

```python
# 배열의 각 요소는 axis 인덱스 배열로 참조할 수 있다.
# 1차원 배열은 1개의 인덱스, 2차원 배열은 2개의 인덱스, 3차원 배열은 3개의 인덱스 요소로 참조할 수 있다.
# 인덱싱 작업시 [행][열]과 같이 인덱스를 지정하거나 [행, 열]과 같이 인덱스를 지정할 수 있다.
```

```python
# 데모 배열 생성
a = np.arange(24) # 1차원 배열
pprint(a)
b = np.arange(24).reshape(4, 6) # 2차원 배열
pprint(b)
c = np.arange(24).reshape(2, 4, 3) # 3차원 배열
pprint(c)
```

```python
# 1차원 배열 요소 참조 및 변경
print(a[5])
a[0] = 1000000
pprint(a)
```

```python
# 2차원 배열 요소 참조 및 변경
print(b[0][1])
print(b[1, 3])
b[0][1] = 1000000
b[1, 3] = 2000000
pprint(b)
```

```python
# 3차원 배열 요소 참조 및 변경
print(c[1][0][1])
print(c[0, 2, 1])
c[1][0][1] = 999999
c[0, 2, 1] = 777777
pprint(c)
```

불린 인덱싱  
numpy의 불린 인덱싱은 배열의 각 요소 선택 여부를 True, False로 지정하는 방식이다.

```python
# 데모 배열 생성
a = np.arange(24).reshape(4, 6)
pprint(a)
```

```python
# a 배열에서 요소의 값이 짝수인 요소의 총 합계
# numpy의 브로드캐스팅 기능을 이용해서 짝수인 배열의 요소를 확인한다.
even_arr = a % 2
pprint(even_arr)
even_arr1 = a % 2 == 0
pprint(even_arr1)
```

```python
# 불린 인덱싱을 이용하면 True에 해당되는 요소들만 인덱싱한다.
print(a[a % 2 == 0])
print(a[even_arr1])
```

```python
print(np.sum(a))
print(np.sum(a[a % 2 == 0]))
print(np.sum(a[a % 2 == 1]))
```

```python
print('1월 강수량 합계 : {}'.format(np.sum(rain_jan)))
print('1월 강수량 평균 : {}'.format(np.mean(rain_jan)))
```

배열 변환  
배열을 변환하는 방법은 전치, 배열 shape 변환, 배열 요소 추가, 배열 결합, 배열 분리 등이 있다.

```python
# 전치는 행렬의 인덱스가 바뀌는 변환이다.
Image('./numpyImage/numpyImage08.jpg', width='250')
```

```python
# 넘파이에서 행렬을 전치하기 위해 배열이름.T 속성을 사용한다.
a = np.random.randint(1, 10, (2, 3))
pprint(a)
pprint(a.T)
```

```python
# 배열 형태 변경
# 넘파이는 배열의 shape을 변경하는 reshape()와 ravel() 메소드를 제공한다.
# reshape() 메소드는 데이터 변경없이 지정된 shape으로 배열의 형태를 변환한다.
# ravel() 메소드는 배열의 shape을 1차원으로 변환한다.
```

```python
a = np.random.randint(1, 10, (2, 3))
pprint(a)
pprint(a.ravel())
pprint(a)
```

```python
b = a.ravel()
pprint(b)
b[0] = 999
pprint(b)
# ravel() 메소드도 인덱싱, 슬라이싱의 경우처럼 원본 배열의 view를 보여준다.
# 따라서, ravel() 메소드가 반환한 배열을 수정하면 원본 배열의 내용이 같이 수정된다.
pprint(a)
```

```python
a = np.random.randint(1, 10, (2, 3))
pprint(a)
b = a.reshape(3, 2, 1)
pprint(b)
b[0][0][0] = 999
pprint(b)
# reshape() 메소드도 인덱싱, 슬라이싱의 경우처럼 원본 배열의 view를 보여준다.
# 따라서, reshape() 메소드가 반환한 배열을 수정하면 원본 배열의 내용이 같이 수정된다.
pprint(a)
```

```python
# resize() 메소드는 배열의 shape과 크기를 변경한다.
# np.resize(배열이름, new_shape), 배열이름.resize(new_shape)
# resize()와 reshape() 메소드는 shape를 변경한다는 부분에서는 유사하지만 reshape() 메소드는 요소의 개수를 변경하지 않는 반면에
# resize() 메소드는 shape을 변경하는 과정에서 요소의 개수를 줄이거나 늘릴 수 있다.
```

```python
a = np.random.randint(1, 10, (2, 6))
pprint(a)
np.resize(a, (6, 2)) # np.resize() 방식은 반드시 실행 결과를 새 배열에 저장시켜 사용해야 한다.
pprint(a)
b = np.resize(a, (6, 2))
pprint(b)
```

```python
a = np.random.randint(1, 10, (2, 6))
pprint(a)
a.resize(6, 2) # 배열이름.resize() 방식을 실행하면 배열의 shape가 변경된다.
pprint(a)
```

```python
# 요소의 개수가 변경되는 resize
a = np.random.randint(1, 10, (2, 6))
pprint(a)
# a.reshape(2, 10) # reshape() 메소드는 원본과 요소의 개수가 다르면 에러가 발생된다.
a.resize(2, 10) # 배열 요소의 개수가 12개에서 20개 늘어나고 늘어난 요소에 0이 채워진다.
pprint(a)
a.resize(2, 4) # 배열 요소의 개수가 20개에서 8개 줄어들고 줄어든 요소에 데이터는 분실된다.
pprint(a)
```
