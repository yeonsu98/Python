```python
import warnings
warnings.filterwarnings('ignore')
from IPython.display import Image
import numpy as np
import pandas as pd
```

데이터에 누락값이 없으면 좋겠지만 누락값은 대부분의 데이터에 존재하고 깔끔한 데이터를 위해 누락값을 처리하는 것은 매우 중요하다. 누락값은 NaN으로 표기되며 누락값을 파이썬에서 사용하려면 넘파이 라이브러리가 필요하다.

```python
# 누락값을 사용하기 위해 넘파이에서 누락값을 불러온다.
from numpy import NaN, NAN, nan
```

```python
# 누락값은 0, ''과 같은 값과는 다른 개념으로 데이터 자체가 없다는 것을 의미한다. => NULL => '같다'라는 개념도 없다.
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
```

```python
# 누락값은 값 자체가 없기 때문에 자기 자신과 비교해도 False가 출력된다.
print(NaN == NaN)
print(NaN == NAN)
print(NaN == nan)
```

```python
# 판다스에서는 누락값을 확인하는 isnull(), isna() 메소드가 있고 누락값이면 True, 누락값이 아니면 False를 리턴한다.
print(pd.isnull(NaN))
print(pd.isnull(0))
print(pd.isnull(''))
print(pd.isna(NaN))
print(pd.isna(0))
print(pd.isna(''))
```

```python
# 반대의 경우(누락값이 아닌 경우)도 검사하는 notnull(), notna() 메소드가 있고 누락값이 아니면 True, 누락값이면 False를 
# 리턴한다.
print(pd.notnull(NaN))
print(pd.notnull(0))
print(pd.notnull(''))
print(pd.notna(NaN))
print(pd.notna(0))
print(pd.notna(''))
```

```python
# Cases_* => 발병 국가, Deaths_* => 사망자 수
ebola = pd.read_csv('./data/country_timeseries.csv')
ebola
```

```python
# 넘파이의 count_nonzero() 메소드는 배열에서 0(False)이 아닌 데이터의 개수를 센다.
print(np.count_nonzero([False, 0, True, 1, 999, NaN]))
# isnull() 메소드는 누락값이면 True, 누락값이 아니면 False를 리턴한다.
print(np.count_nonzero(ebola.isnull())) # 누락값의 개수를 센다.
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

```python
# count() 메소드로 누락값이 아닌 데이터의 개수를 셀 수 있다.
print(ebola.count())
```

```python
print(ebola.shape)
print('ebola 데이터는 {}행 {}열 입니다.'.format(ebola.shape[0], ebola.shape[1]))
# 전체 데이터의 개수에서 누락값이 아닌 데이터의 개수를 빼면 누락값의 개수를 알 수 있다.
num_row = ebola.shape[0]
num_missing = num_row - ebola.count()
print(num_missing)
```

```python
# 시리즈의 value_counts() 메소드는 지정한 열의 빈도수를 얻어올 수 있다.
# print(type(ebola['Cases_Guinea']))
# print(ebola['Cases_Guinea'].value_counts())
print(type(ebola.Cases_Guinea)) # 데이터프레임에 '.'을 찍어서 특정 열의 데이터를 얻어올 수 있다. => 시리즈
print(ebola.Cases_Guinea.value_counts())
```

```python
# value_counts() 메소드에 dropna=False 옵션을 지정하면 NaN의 개수도 센다.
print(ebola.Cases_Guinea.value_counts(dropna=False))
```

```python
# 기초 통계를 계산하는 메소드는 skipna=True 옵션이 기본값으로 사용되기 때문에 누락값을 무시하고 메소드가 실행된다.
print(ebola.Cases_Guinea.sum())
print(ebola.Cases_Guinea.sum(skipna=True))
print(ebola.Cases_Guinea.sum(skipna=False))
print(ebola.Cases_Guinea.mean())
print(ebola.Cases_Guinea.max())
print(ebola.Cases_Guinea.min())
print(ebola.Cases_Guinea.count())
```

누락값 처리하기

```python
print(ebola.iloc[:10, :5])
```

```python
# fillna() 메소드의 인수로 특정 데이터를 지정하면 누락값을 지정된 값으로 변경한다.
print(ebola.fillna(0).iloc[:10, :5])
```

```python
# fillna() 메소드에 method='ffill' 옵션을 지정하면 누락값이 나타나기 전의 값으로 변경한다.
print(ebola.fillna(method='ffill').iloc[:10, :5])
```

```python
# fillna() 메소드에 method='bfill' 옵션을 지정하면 누락값이 나타난 다음 값으로 변경한다.
print(ebola.fillna(method='bfill').iloc[:10, :5])
```

```python
# interpolate() 메소드는 누락값 양쪽에 있는 데이터를 이용해서 중간값(평균값)으로 누락값을 변경한다.
print(ebola.interpolate().iloc[:10, :5])
```
