```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도위에 시각화 할 수 있다.
# !pip install folium
import folium
# 보고자 하는 위치의 좌표계를 얻어내는 가장 단순한 방법은 구글맵을 이용하는 방법이다.
```

```python
# folium 라이브러리의 Map() 메소드의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 옵션으롤 지정한 배율을
# 사용하는 지도를 만든다.
# folium.Map(location=[위도, 경도], zoom_start=배율[, tiles='지도 스타일'])
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=17)

# folium의 save() 메소드로 작성된 지도 차트를 html 파일로 저장할 수 있다.
map_folium.save('./output/map.html')
map_folium
```

```python
# 지도 스타일 => Open Street Map(기본값), Stamen Terrain, Stamen Toner
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=17, tiles='Open Street Map')
map_folium
```

```python
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=16)

# folium 라이브러리의 Marker() 메소드의 location 속성으로 지정한 위치에 마커를 만든다.
# popup 속성으로 마커를 클릭했을 때 표시할 문자열을 지정할 수 있다.
# 작성된 마커는 add_to() 메소드를 사용해서 지도에 추가한다.

folium.Marker(location=[37.56652479270633, 126.9779129116262], popup='서울특별시청').add_to(map_folium)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup='덕수궁').add_to(map_folium)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup='서울지방경찰청').add_to(map_folium)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup='광화문').add_to(map_folium)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup='영풍문고 - 종로본점').add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=16)

# 마커를 클릭했을 때 팝업되는 말풍선의 크기가 마음에 들지 않으면 Popup() 메소드의 max_width 속헝으로 변경하면 된다.
popup = folium.Popup('서어어어어어우우우우우울트으으으으윽벼어어어얼시이이이이이처어어어어어엉', max_width=500)

folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup).add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```

```python
# icon 속성에 Icon() 메소드로 마커의 색상과 아이콘 모양을 지정할 수 있다.
# Icon() 메소드의 color 속성으로 지정할 수 있는 색상은 아래와 같다.
# white', 'lightred', 'green', 'pink', 'orange', 'beige', 'darkred', 'darkgreen', 'cadetblue', 'darkpurple', 'lightgray', 
#'purple', 'black', 'blue', 'red', 'lightblue', 'gray', 'darkblue', 'lightgreen'
# 마커 위에 표시되는 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서 제공하는 이미지를 사용할 수
# 있다.

map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=16)

folium.Marker(location=[37.56652479270633, 126.9779129116262], popup='서울특별시청',
             icon=folium.Icon(color='beige', icon='star')).add_to(map_folium)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup='덕수궁',
             icon=folium.Icon(color='green', icon='ok')).add_to(map_folium)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup='서울지방경찰청',
             icon=folium.Icon(color='blue', icon='user')).add_to(map_folium)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup='광화문',
             icon=folium.Icon(color='orange', icon='hand-down')).add_to(map_folium)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup='영풍문고 - 종로본점',
             icon=folium.Icon(color='purple', icon='eye-open')).add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=16)

# folium 라이브러리의 CircleMarker() 메소드로 원 모양의 마커를 만들 수 있다.
folium.CircleMarker(location=[37.56652479270633, 126.9779129116262], popup=popup,
                   radius=100, color='#12A57A', fill_color='#E557A7').add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```
